```python
no_of_primary_tech = df_2020[primary_tech].value_counts()
```

```python
no_of_primary_tech
```

```python
top_10_primary_tech = no_of_primary_tech[:10]
top_10_primary_tech
```

```python
plt.figure(figsize=(10,6))
plt.ylim(0,200)
plt.xticks(rotation=45)
plt.bar(top_10_primary_tech.index,top_10_primary_tech.values)
plt.show()
```

```python
primary_tech_dict = {}
for item in no_of_primary_tech.iteritems():
    keys = item[0].replace('/',',').split(',')
    keys = [key.strip().lower() for key in keys]
    value = item[1]
    for key in keys:
        if key not in primary_tech_dict:
            primary_tech_dict[key] = 0
        primary_tech_dict[key]+=value
#     primary_tech_dict.append({key:item[1]})
```

```python
# primary_tech_dict
```

```python
new_top_10 = pd.Series(primary_tech_dict)
```

```python
sorted_top_10 = new_top_10.sort_values(ascending=False)[:10]
```

```python
sorted_top_10
```

```python
plt.figure(figsize=(10,6))
plt.xticks(rotation=45)
plt.ylim(0,260)
plt.yticks(ticks = [x for x in range(0,251,25)],labels = [x for x in range(0,251,25)])
plt.bar(sorted_top_10.index,sorted_top_10.values)
plt.show()
```
